### Cleaning dataset

We're in presence of a panel dataset. We choose to only select observations for the year `2012`. 
We're also gonna convert the `Adult Mortality` continuous variable into a categorical variable.

```python
df = pd.read_csv("dataset/life-expectancy.csv", sep=",", decimal=".")
print(df.shape)
print(df.columns)
df.head()
```

```python
df.rename(str.strip, axis=1, inplace=True)
df.columns = df.columns.str.lower()
df.rename({"thinness  1-19 years": "thinness 10-19 years"}, axis=1, inplace=True)
df["thinness  1-19 years"].rename("thinness 10-19 years", inplace=True)

print(df.columns)
df.head()
```

```python
df = df[df["year"] == 2012]
print(df.shape)
df.head()
```

```python
df["adult mortality"] = pd.cut(df["adult mortality"], [0, 50, 80, 150, 350, df["adult mortality"].max()], labels=["very low", "low", "middle", "high", "very high"])
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
for feature in df.columns:
    df.dropna(subset=[feature], inplace=True)
    
df.drop("year", axis=1, inplace=True)

df.head()
```

```python
df.shape
```

```python
df.to_csv("dataset/life-expectancy-2012.csv", index=False)
```
